```python
import os
from tqdm import tqdm
import shutil
import glob
```

```python
#directory with raw unprocessed data
RawData_Directory = r'C:\Users\felix\Desktop\DAS\mrlEyes_2018_01\mrlEyes_2018_01'

#copy the images with 0 state to closed eyes folder and images with 1 state to opened eyes folder
for dirpath ,dirname, filenames in os.walk(RawData_Directory):
    for i in tqdm([m for m in filenames if m.endswith('.png')]):
        #print(i)
        
        if i.split('_')[4]=='0':
            shutil.copy(src = dirpath + '/' + i, dst = r'C:\Users\felix\Desktop\DAS\mrlEyes_2018_01\sorted data\closed eyes')
            
        elif i.split('_')[4]=='1':
            shutil.copy(src = dirpath + '/' + i, dst = r'C:\Users\felix\Desktop\DAS\mrlEyes_2018_01\sorted data\opened eyes')
            
        
    
```
